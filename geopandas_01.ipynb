```python
#importamos la libreria
import geopandas as gpd
```

```python
#leemos el shapefile
#esta forma de leer el shapefile es cuando el archivo esta en otra carpeta

localidades_shp = gpd.read_file(r"C:\Users\Seba\Documents\geo_pandas\departamentos\Departamentos.shp")
# imprimimos el shapefile
localidades_shp
```

```python
#ponemos el nombre de la localidad como index
localidades_shp = localidades_shp.set_index("NOMBRE")
localidades_shp
```

```python
#ordenamos alfabeticamente por nombre
localidades_shp.sort_values("NOMBRE")
```

```python
#ordenamos por superficie
localidades_shp.sort_values("SUPERFICIE")
```

```python
#valor maximo
localidades_shp["SUPERFICIE"].max()
```

```python
#valor minimo
localidades_shp["SUPERFICIE"].min()
```

```python
#valor medio
localidades_shp["SUPERFICIE"].mean()
```

```python
#visualizamos el mapa utilizando la superficie como parametro
localidades_shp.plot("SUPERFICIE", legend = True)
```

```python
# Graficamos solo el Departamento Rio Cuarto
rio_cuarto = localidades_shp.loc["Río Cuarto", "geometry"]
rio_cuarto
```
